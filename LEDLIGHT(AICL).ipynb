```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import pandas as pd
import numpy as np
import time
```

```python
#PATIENCE_TIME = 60
#LOAD_MORE_BUTTON_XPATH = '//*[@id="browse-itemsprimary"]/li[2]/button/span/span[2]' 

#driver = webdriver.Chrome('./chromedriver.exe')
#driver.get(YOUTUBER_HOME_PAGE_URL)
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://www.exportersindia.com/search.php?srch_catg_ty=prod&term=LED&cont=IN&ss_status=N')
while True:
    try:
        loadMoreButton = driver.find_element_by_xpath("//a[contains(@class,'dib p7px w170px bgfff gray ttu bdr darkbdr1 br5px')]")
        time.sleep(2)
        loadMoreButton.click()
        time.sleep(5)
    except Exception as e:
        print (e)
        break
print ("Complete")
time.sleep(10)
#driver.quit()
```

```python
lists = []

for i in range (1,2):
    name = driver.find_elements(By.XPATH, '//p[@class="mb5px ffos xlarge graydark b dib"]')
    no = driver.find_elements(By.XPATH, '//a[contains(@class,"view_mobile number") or contains(@class,"view_mobile")]')
    ad = driver.find_elements(By.XPATH, '//span[@class="black title_tooltip"]')
    #//td[contains(@class,'blob-code-addition') or contains(@class,'blob-code-deletion')]
    #no = driver.find_elements(By.XPATH, "//a[contans(@class, 'view_mobile number') or contains(@class, 'view_mobile')]")
    
    for i in range(len(name)):
        lists.append([name[i].text, no[i].text, ad[i].text])
        
    time.sleep(10)    
```

```python
data = pd.DataFrame(lists, columns = ['Name', 'Number', 'Address'])
```
