```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
AI = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
AI.status_code
```

```python
soup = BeautifulSoup(AI.text, 'html.parser')
```

```python
title = soup.find_all('h2', class_ = 'sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg') 
title = pd.DataFrame([x.text for x in title], columns = ['Paper_Title'])
```

```python
author = soup.find_all('span', class_ = 'sc-1w3fpd7-0 dnCnAO')
author = pd.DataFrame([x.text for x in author], columns = ['Authors'])
```

```python
date = soup.find_all('span', class_ = 'sc-1thf9ly-2 dvggWt')
date=pd.DataFrame([x.text for x in date], columns = ['Published_Date'])
```

```python
links = pd.DataFrame([link.get('href') for link in soup('a', class_ = 'sc-5smygv-0 fIXTHm')], columns = ['Paper_url'])
```

```python
AI_data = pd.concat([title, author, links], axis = 1)
AI_data
```
