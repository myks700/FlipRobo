```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from lxml import etree as et
```

```python
president = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
print(president.status_code)
soup= BeautifulSoup(president.text, 'html.parser')
```

```python
Name = soup.find_all('h3')
Name = [i.text.strip() for i in Name]
Name = pd.DataFrame(Name, columns = ['Name'])
Name
```

```python
terms = soup.find_all('div', class_ = 'presidentListing')
dt = [x.text.strip() for x in terms]
dt = pd.DataFrame(dt, columns = ['Name'])
data =  dt['Name'].str.split('\n', expand=True)
```

```python
data = data.rename(columns = {0 : 'Name', 1: 'Terms'})
data = data.iloc[:,0:2]
data
```

```python
Terms = data['Terms'].str.split(':', expand=True)
Terms = Terms.iloc[:,1:2].rename(columns = {1:'Terms'})

data = pd.concat([Name, Terms], axis=1, join="inner")
data
```
